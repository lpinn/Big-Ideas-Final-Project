# Import statements and reading the playgrounds csv

```python
import pandas as pd
import matplotlib.pyplot as plt

playgrounds = pd.read_csv("https://data.wprdc.org/datastore/dump/47350364-44a8-4d15-b6e0-5f79ddff9367")
#playgrounds.head(5)
```

# Creating the metric function, reading each neighborhood and keeping count of how many times it appears, 1 appearance is 1 point

```python
playgroundsDictionary = {}

names = []
names = playgrounds['neighborhood'].unique()

length_of_names = len(names)-1

def playgroundsMetric():
    counter = 0
    while(counter < length_of_names):
        for i in range(len(playgrounds)-1):
            if(playgrounds['neighborhood'].iloc[i] is names[counter]):
                if(names[counter] in playgroundsDictionary):
                    playgroundsDictionary[playgrounds['neighborhood'].iloc[i]] += 1    
                else:
                    playgroundsDictionary[playgrounds['neighborhood'].iloc[i]] = 1      
        counter += 1

playgroundsMetric()

df = pd.DataFrame.from_dict(playgroundsDictionary, orient="index", columns=['Points'])
print(playgroundsDictionary)
```

# Visualizations

```python
values = list(playgroundsDictionary.values())
labels = list(playgroundsDictionary.keys())

fig = plt.figure(figsize = (20, 5))

plt.xticks(rotation=90)

plt.bar(labels, values)
```

```python
df.sort_values(ascending=False, by="Points")
```

```python
df["Points"].plot.box()
```
