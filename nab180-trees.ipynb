```python
import pandas as pd
import numpy as np
```

```python
treesData = pd.read_csv("https://data.wprdc.org/datastore/dump/1515a93c-73e3-4425-9b35-1cd11b2196da")
dictionaryMetric = {}
print(treesData.head(1))
```

```python
def treeMetricAdd():
    for i in range(0, len(treesData) - 1):
        if (treesData["neighborhood"].iloc[i] in dictionaryMetric):
            dictionaryMetric[treesData["neighborhood"].iloc[i]] += 1
        else:
            dictionaryMetric[treesData["neighborhood"].iloc[i]] = 1
            
treeMetricAdd()
print(dictionaryMetric)
```
