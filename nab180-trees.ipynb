## Import packages 

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

## Import trees dataset (into pandas)

```python
treesData = pd.read_csv("https://data.wprdc.org/datastore/dump/1515a93c-73e3-4425-9b35-1cd11b2196da", index_col="id", low_memory=False)
dictionaryMetric = {}
print(treesData.head(1))
```

## Create a tree metric. 10 trees = 1 point.

```python
def treeMetricAdd():
    for i in range(0, len(treesData) - 1):
        if (treesData["neighborhood"].iloc[i] in dictionaryMetric):
            dictionaryMetric[treesData["neighborhood"].iloc[i]] += 0.1
        else:
            dictionaryMetric[treesData["neighborhood"].iloc[i]] = 0.1
    for k in dictionaryMetric:
        dictionaryMetric[k] = round(dictionaryMetric[k])
            
treeMetricAdd()
print(dictionaryMetric)
```

## Graph for analysis

```python
n = list(dictionaryMetric.keys())
val = list(dictionaryMetric.values())
plt.figure(figsize=(20, 30))
plt.barh(range(len(dictionaryMetric)),val,tick_label=n)
```

## Conclusion

I chose my metric to be 1 point for every 10 trees. The reason I did this was to make the amount of trees scale reasonably to my team mates' metrics. I believe 10 trees is about equivilant to a small inner-city park in terms of enjoyment. The data set I used was the "trees" data set by WPRDC, which includes the location of every tree in inner pittsburgh, along with which neighborhood it resides in. After analyzing the data, I found that Squirrel Hill South contained the greatest amount of trees, accumilating 507 points. This makes sense because Squirrel Hill South contains Schenley and Frick park. Highland park has the second greatest amount of trees, accumilating 294 points. This also makes sense, considering that half of Highland park is a park.
